# 데이터 전처리

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
%matplotlib inline
```

```python
df1 = pd.read_excel('./여행지별_해시태그.xlsx')
```

```python
df1['tags'] = df1['tags'].str.replace('#', '')
```

```python
df1
```

```python
df1.shape
```

```python
tag_iter = (set(x.split(',')) for x in df1.tags)
tag_set = sorted(set.union(*tag_iter))

tag_set
```

```python
tag_mat = pd.DataFrame(np.zeros((len(df1), len(tag_set))), columns=tag_set)

tag_mat
```

```python
for i, genre in enumerate(df1.tags):

              tag_mat.loc[i, genre.split(',')] = 1

x = tag_mat
y = df1['place']
```

```python
x.shape
```
